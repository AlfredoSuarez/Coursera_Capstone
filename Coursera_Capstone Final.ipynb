```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.utils import resample
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data=pd.read_csv("https://s3.us.cloud-object-storage.appdomain.cloud/cf-courses-data/CognitiveClass/DP0701EN/version-2/Data-Collisions.csv")
```

### Data understanding:
As you can notice, this dataset consists of 37 attributes; which can be splitted between independent variables and the 'SEVERITYCODE' as dependent variable(target). The objective is build a model to predict the 'SEVERITYCODE' based on those independent variables(note:we might exclude some of them due significance).
<ul>
    <li>To start some issues need to be fixed:
    <li>It has too many attributes (37)</li>
    <li>It is an unbalanced dataset (SEVERITYCODE)</li>

</ul>

Considering the facts already shown, for the study would be selected as independent variables, Road Conditions, Liight Conditions, Weather and Speeding. These variables could provide major signifiance of probabilities of an accident may occur and how badly may resault(to predict SEVERITYCODE variable).

The first step is to select the wanted attributes:

```python
df=pd.DataFrame(data[['SEVERITYCODE','WEATHER','ROADCOND', 'LIGHTCOND','SPEEDING']])
df.head()
```

Next is to fill the Nan values, in order to place them through the study:

```python
df["SPEEDING"].fillna("N", inplace = True)
df.head()
```

Then, the raw data has different characteristics, so in oorder to place a common scale, is necessary to transform it, this case would be through sequencial numbers like shown below

```python
encoding_weather={'WEATHER':{'Clear': 1, 'Partly Cloudy':2, 'Overcast':3, 'Fog/Smog/Smoke':4, 'Severe Crosswind':5, 'Raining':6,'Snowing':7,'Sleet/Hail/Freezing Rain':8,'Blowing Sand/Dirt':9,'Unknown':10, 'Other':11}}
df.replace(encoding_weather, inplace=True)
df.head()
```

```python
encoding_roadcond={'ROADCOND':{'Dry': 1, 'Sand/Mud/Dirt':2, 'Oil':3, 'Wet':4, 'Standing Water':5, 'Snow/Slush':6, 'Ice':7,'Other':8,'Unknown':9}}
df.replace(encoding_roadcond, inplace=True)
df.head()
```

```python
encoding_lightcond={'LIGHTCOND':{'Daylight': 1, 'Dawn':2, 'Dusk':3, 'Dark - Street Lights On':4, 'Dark - Street Lights Off':5,'Dark - No Street Lights':6,'Dark - Unknown Lighting':7,'Other':8,'Unknown':9}}
df.replace(encoding_lightcond, inplace=True)
df.head()
```

```python
encoding_speed={'SPEEDING':{'Y':2,'N':1}}
df.replace(encoding_speed, inplace=True)
df.head()
```

After that, lets balance out the dataset of SEVERITYCODE with 1 and 2 characters:

```python
data_max= data[data.SEVERITYCODE==1]
data_min= data[data.SEVERITYCODE==2]

data_max_upsampled=resample(data_max,
                           replace=True,
                           n_samples=58188,
                           random_state=123)

balanced_df=pd.concat([data_max_upsampled, data_min])
balanced_df.SEVERITYCODE.value_counts()
```

After the Nan values or Null will be removed

```python
df.dropna(axis='rows',inplace=True)
```

```python
df.isnull().sum()
```

Now We can create an array with the new scale to study the data with dependent and independent variables, in this case lets use a matrix aproach:

```python
x= np.asarray(df[['WEATHER', 'ROADCOND','LIGHTCOND','SPEEDING']])
x[0:5]
```

```python
y=df['SEVERITYCODE'].values
y[0:5]
```

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(x).transform(x)
X[0:5]
```

As the Matrix has being placed, now lets split the dataset, 75% for trainning, 25% testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=0.25, random_state=4)

print('Train set shape: ', X_train.shape,y_train.shape)
print('Test set shape: ', X_test.shape,y_test.shape)
```

### Models
This case, lets see different models behave with the dataset and can help to estimate the Severity of an accident. The models are:
<ul>
    <li>KNN</li>
    <li>Decision Tree</li>
    <li>SVM</li>
    <li>Logistic Regression</li>
</ul>


For the K Neighbor Model, lets try first with a number, this case 4; afterwards lets check it is an optimal point.

```python
#KNN

from sklearn.neighbors import KNeighborsClassifier
k=4
neigh =KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train)
neigh

Kyh= neigh.predict(X_test)
Kyh

print("Train set accuracy= ", metrics.accuracy_score(y_train, neigh.predict(X_train)))
print("Test set accuracy= ", metrics.accuracy_score(y_test, Kyh))
```

```python
#Best KNN
Ks= 20

mean_acc = np.zeros((Ks-1))
std_acc=np.zeros((Ks-1))
ConfusionMx=[];

for n in range(1,Ks):
    neigh=KNeighborsClassifier(n_neighbors = n).fit(X_train, y_train)
    yh=neigh.predict(X_test)
    mean_acc[n-1]=metrics.accuracy_score(y_test,yh)
    std_acc[n-1]=np.std(yh==y_test)/np.sqrt(yh.shape[0])

mean_acc
```

And it was not! Usually is necessary to becarefull on not overfitting model and not forced the model to be a high number, though by the size of data set, and the time that may take to do more testing in the near future for more studies, in a debatle way lets take the number K=16 as good number. Nevertheless lets see how good this K number is by evaluating the model

```python
#Decision Tree

from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier(criterion="entropy", max_depth =5)
DT.fit(X_train,y_train)

DTyh=DT.predict(X_test)

print("Train set accuracy= ", metrics.accuracy_score(y_train, DT.predict(X_train)))
print("Test set accuracy= ", metrics.accuracy_score(y_test, DTyh))
```

```python
#SVM
from sklearn import svm

clf= svm.SVC(kernel='rbf')
clf.fit(X_train, y_train)

SVMyh=clf.predict(X_test)

print("Train set accuracy= ", metrics.accuracy_score(y_train, clf.predict(X_train)))
print("Test set accuracy= ", metrics.accuracy_score(y_test, SVMyh))
```

```python
#Logistic Regression

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix


LR= LogisticRegression(C=6,solver='liblinear').fit(X_train, y_train)

LRyh=LR.predict(X_test)
y_proba=LR.predict_proba(X_train)
print("Train set accuracy= ", metrics.accuracy_score(y_train, LR.predict(X_train)))
print("Test set accuracy= ", metrics.accuracy_score(y_test, LRyh))
```

### Results
According to the findings, We have got the following:
The best number for K Neighbor model would be K=4 due has ma better Jaccard and F1 score. Also, the resaults table the best models for future analysis with this dataset would be KNN and logistic regression if involving a binary answers for future questions that may occur. Also, needs to bring more attention on how significant would be for "car accidents" the "SPEEDING" variable. May help to kill the myth that high speed is bad?


```python
#Model Evaluation

from sklearn.metrics import f1_score
from sklearn.metrics import jaccard_similarity_score
from sklearn.metrics import log_loss

print('KNN jaccard index: %.7f' % jaccard_similarity_score(y_test,Kyh))
print('KNN f1_score index: %.7f' % f1_score(y_test,Kyh, average='weighted'))

print('Decision Tree jaccard index: %.7f' % jaccard_similarity_score(y_test,DTyh))
print('Decision Tree f1_score index: %.7f' % f1_score(y_test,DTyh, average='weighted'))

print('SVM jaccard index: %.7f' % jaccard_similarity_score(y_test,SVMyh))
print('SVM f1_score index: %.7f' % f1_score(y_test,SVMyh, average='weighted'))

print('LogReg jaccard index: %.7f' % jaccard_similarity_score(y_test,LRyh))
print('LogReg f1_score index: %.7f' % f1_score(y_test,LRyh, average='weighted'))

print('LogReg logLoss: %.7f' %log_loss(y_train,y_proba))
```
